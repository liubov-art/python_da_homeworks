7.2 Сформулюйте 3 гіпотези, чому в одному місті gross income вище, як можна дізнатись чи правдива ця гіпотеза з використанням даних (наявних і додаткових), та як висновки з цих гіпотез можна використати для прийняття рішень як підвищити прибуток в мережі супермаркетів. Відповідь напишіть текстом в окремій клітинці типу Markdown.

Це завдання - на продуктове мислення. Якщо не знаєте, які гіпотези сформулювати - напишіть ті, які здаються вам зовсім примітивними. Важливо навчитись висловлювати свої гіпотези щодо даних, і з часом та досвідом вони будуть ставати щоразу глибше і точніше. Весь аналіз даних будується на тому, що ми перевіряємо певні гіпотези та робимо висновки і рекоемндації на основі своїх досліджень.

```python
mean_yangon = df[df.City =='Yangon'] ["gross income"].describe()["mean"]
mean_mandalay = df[df.City =='Mandalay'] ["gross income"].describe()["mean"]
print(f"Cереднє значення gross income для Yangon = {mean_yangon}, cереднє значення gross income для Mandalay = {mean_mandalay}.")

if mean_yangon >  mean_mandalay:
  print(f"Cереднє значення gross income більше в місті Yangon")
else:
  print(f"Cереднє значення gross income більше в місті Mandalay")
```

9. Врахуовуючи висновок в попередньому завданні, дослідимо як продаються в цих містах різні категорії товарів. Виведіть відсотковий розпоіділ кількості інвойсів по категоріям товарів `Product line` в містах `Yangon` та `Mandalay`.

```python
product_line_yangon_count = df[df.City =='Yangon']["Product line"].value_counts()
percentage_prod_line_yangon = product_line_yangon_count / product_line_yangon_count.sum()
product_line_mandalay_count = df[df.City =='Mandalay']["Product line"].value_counts()
percentage_prod_line_mandalay = product_line_mandalay_count / product_line_mandalay_count.sum()
print(f"Відсотковий розпоіділ кількості інвойсів по категоріям товарів Product line в місті Yangon: {percentage_prod_line_yangon}")
print(f"Відсотковий розпоіділ кількості інвойсів по категоріям товарів Product line в місті Madalay: {percentage_prod_line_mandalay}")
```
