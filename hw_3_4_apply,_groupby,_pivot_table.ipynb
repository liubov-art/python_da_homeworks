**1.Чи є різниця в популярності продуктів серед різних типів клієнтів (члени та звичайні покупці) в різних містах. Та чи впливає гендер на тип продукту в різних містах.**

В таблиці city_product_customer можна побачити скільки продуктів кожного типу було придбано різними категоріми клієнтів - Member та Normal. Також можна відслідкувати, які типи продуктів більше купували жінки, а які чоловіки в певних містах.

```python
city_product_customer = pd.pivot_table(
    df,
    values='Quantity',
    index=['City', 'Product line'],
    columns=['Customer type','Gender'],
    aggfunc='sum'
)
city_product_customer
```

Наступна таблиця показує, які категорії товарів зустрічаються найчастіше в різних містах для різних типів клієнтів

```python
def get_mode(series):
  return series.mode().iloc[0]
```

```python
pivot_table = pd.pivot_table(
    df,
    values='Product line',
    index=['Customer type', 'Gender'],
    columns='City',
    aggfunc=get_mode,
)
pivot_table
```

Наступна таблиця показує які категорії товарів купували найчатіше в різних містах, різні типи покупців (групувала по Quantity). Дані відрізняються від попередньї таблиці, бо в колонці 'Product line' категорія товарів зазначена лише 1 раз, а в Quantity рахується кількість товарів, як були куплені в даній категорії

```python
def get_max_product(group):
  quantity_prodline = group.groupby('Product line')['Quantity'].sum()
  return quantity_prodline.idxmax()
```

```python
pivot_table_product = df.groupby(['Customer type', 'Gender', 'City']).apply(get_max_product).unstack()
pivot_table_product
```

В наступній таблиці я порахувала кількість продуктів (по категоріям), які буди куплені найчастіше

```python
def get_max_quantity(group):
  quantity_prodline = group.groupby('Product line')['Quantity'].sum()
  return quantity_prodline.max()
```

```python
pivot_table_quantity = df.groupby(['Customer type', 'Gender', 'City']).apply(get_max_quantity).unstack()
pivot_table_quantity
```

Ця таблиця показує об'єднані значення

```python
concatenated_df = pd.concat([pivot_table_product, pivot_table_quantity], keys=['Категорія товарів', 'Кількість товарів'], axis=1)
concatenated_df
```

Тут також об'єднані значення, але в іншому форматі.

```python
def get_max_product_and_quantity(group):
    quantity_prodline = group.groupby('Product line')['Quantity'].sum()
    max_product = quantity_prodline.idxmax()
    max_quantity = quantity_prodline.max()
    return pd.Series([max_product, max_quantity], index=['Категорія товарів', 'Кількість товарів'])
```

```python
pivot_table_product_quantity = df.groupby(['Customer type', 'Gender', 'City']).apply(get_max_product_and_quantity)
pivot_table_product_quantity
```

**2.Чи є певні місяці або дні тижня, коли клієнти купують більше товарів, і чи це пов'язано з сезоном або святами**

```python
def get_date(date):
  return int(date.split('/')[1])
```

```python
def get_month(date):
  return int(date.split('/')[0])
```

```python
df['Day'] = df.Date.apply(get_date)
df['Month'] = df.Date.apply(get_month)
```

```python
df.head()
```

```python
day_quantity_pivot_table = pd.pivot_table(
    df,
    values='Quantity',
    index=['Month', 'Day'],
    columns= ['City', 'Product line'],
    aggfunc='sum'
)
day_quantity_pivot_table.fillna(0)
```

```python
city_day = day_quantity_pivot_table.stack().idxmax()
max_value = day_quantity_pivot_table.stack().max()
```

```python
max_day_quantity = pd.DataFrame({"Max value": max_value, "Month / Day / Product" : city_day})
max_day_quantity
```

```python
day_total_pivot_table = pd.pivot_table(
    df,
    values='Total',
    index=['Month', 'Day' ],
    columns= ['City', 'Product line'],
    aggfunc='sum'
)
day_total_pivot_table.fillna(0)
```

```python
city_day_total = day_total_pivot_table.stack().idxmax()
max_value_total = day_total_pivot_table.stack().max()
```

```python
max_day_total = pd.DataFrame({"Max Total": max_value_total, "Month / Day / Product" : city_day_total})
max_day_total
```

```python
concatenated_df_city_day = pd.concat([max_day_quantity, max_day_total], axis=1)
concatenated_df_city_day
```
